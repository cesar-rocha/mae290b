# Define function to test RK3 code

### The system to solve is  \begin{align} \dot x_1 = \sin (\omega t)  \\ \dot x_2 = \sin (2\omega t)  \\ \dot x_3 = \sin (3 \omega t) \\ \end{align}

```python
def func3(t,x,omg1=1.,omg2=2.,omg3=3.):
    '''
    A function array to a test problem
    '''
    x = np.zeros(3)
    x[0],x[1],x[2] = np.sin(omg1*t),np.sin(omg2*t),np.sin(omg3*t)
    return x
```

# Define functions to integrate a system of ODEs

```python
def my_ode(F,X0,t0,tf,dt,scheme='rk3'):
    
    '''
    Integrates dX/dt = F(X,t), where X and F 
    are N-dimensional arrays using RK3. 
    - t0/tf : initial/final time
    - X0 : initial value
    - dt : time step
    
    returns solution (X) and time array (t)
                                        '''
    try:
        N=X0.size
    except AttributeError:
        N=1
    
    t = np.arange(t0,tf+dt,dt)
    X = np.zeros((N,t.size))
    X[:,0] = X0
    
    for i in range(1,t.size):
        X[:,i] = stepforward(X[:,i-1],F,t[i-1],dt,scheme)
        
    return X,t

def stepforward(X,F,t,dt,scheme='rk3'):
    
    '''
    It steps forward a system of first-order ODEs using
    dX/dt = F(X,t), where X and F are N-dimensional
    arrays.
    
    NOTE: nly Explicit Euler (EE) and third-order 
          Runge-Kutta (RK3) schemes are implemented
                                                    '''
    if scheme=='rk3':
        f1 = F(t,X)
        a1 = 8/15.
        X1 = X + dt*a1*f1
        f2 = F(t+a1*dt,X1)
        a2 = 2/3.
        X2 = X + dt*a2*f2
        f3 = F(t+a2*dt,X2)
        Xnew = X + dt/4.*(f1+3.*f3)
    elif scheme=='ee':
        Xnew = X + dt*F(t)
            
    return Xnew

```

# Compute exact solution to the test problem

### The absolute error is defined as \begin{equation} Error = \frac{1}{3} \sum_{i=1}^{3} |x^{exa}_i - x^{num}_i| \end{equation}

```python
def exact_soln(x0,t,omg1=1.,omg2=2.,omg3=3.):
    ''' Compute exact solution to dx/dt=sin(omg*t) 
        with x(t=0) = x0'''
    xe = np.zeros((3,t.size))
    xe[0,:] = x0[0] + (1.-np.cos(omg1*t))/omg1
    xe[1,:] = x0[1] + (1.-np.cos(omg2*t))/omg2
    xe[2,:] = x0[2] + (1.-np.cos(omg3*t))/omg3
    
    return xe

def residual(x,x_num):
    ''' Compute absolute error in x_num as 
        mean(|x-x_num|)'''
    return np.abs(x-x_num).mean(axis=-1)
```

```python
te = np.arange(0,1000.1,.1)
x_exact = exact_soln(x0,te)
```

# Define functions (RHS) of the Lorenz system

### The Lorenz system is  \begin{align} \dot x = \sigma (y - x),\\ \dot y = \rho x - y - x z,\\ \dot z = -\beta z + x y, \end{align} where $\sigma$, $\rho$, and $\beta$ are parameters. In this problem, we fix $\sigma = 10$ and $\beta = 8/3$, and compute solutions with $\rho = 20$ and $\rho = 28$.

```python
def func_lorenz(t,X,rho=20,sigma=10.,beta=8/3.):
    '''
    Function (RHS) of the Lo
    renz system
    X and F are 3 by 1 arrays.
    '''
    F = np.zeros(X.size)
    F[0] = sigma*(X[1]-X[0])
    F[1] = rho*X[0] - X[1] - X[0]*X[2]
    F[2] = -beta*X[2] + X[0]*X[1]
    
    return F

def jacob_lorenz(X,rho=20,sigma=10.,beta=8/3.):
    '''
    Jacobian matrix for the system in problem 1:
    Lorenz system
    '''
    J = np.zeros((X.size,X.size))
    J[0,0],J[0,1],J[0,2] = -sigma,sigma,0
    J[1,0],J[1,1],J[1,2] = rho-X[2],-1.,-X[0]
    J[2,0],J[2,1],J[2,2] = X[1],X[0],beta
    
    return J

X0 = np.array([1.,1.,1.])
J0 = jacob_lorenz(X0,rho=20)
print J0
lam = np.linalg.eigvals(J0)
lam_abs = np.abs(lam)
print "lam_max/lam_min = " + str(lam_abs.max()/lam_abs.min())
print "max. dt for stability: " + str(2.51/lam_abs.max())
      
```

# Problem 2 – Chemical reactions

### (b) Compute the eigenvalues of the Jacobian matrix at time

```python
import scipy as sp
import scipy.integrate
import scipy.linalg
import time
```

```python
k1,k2,k3=0.04,10.0,1.5e3
C0 = np.array([.9,.1,0.])

def jacob_chem(X,k1=0.04,k2=10.0,k3=1.5e3):
    '''
    Jacobian matrix for the system in problem 2:
    chemical reactions
    '''
    J = np.zeros((X.size,X.size))
    J[0,0],J[0,1],J[0,2], = -k1,k2*X[2],k2*X[1]
    J[1,0],J[1,1],J[1,2] = k1,-k2*X[2]-4.*k3*X[1],-k2*X[1]
    J[2,1] = 4.*k3*X[1]
    
    return J

J0 = jacob_chem(C0)
lam = np.linalg.eigvals(J0)
print lam
lam_abs = np.abs(lam)
print "lam_max/lam_min = " + str(lam_abs.max()/lam_abs.min())

print "max. dt for stability: " + str(2.51/lam_abs.max())
    
```

### (c) Solve the system with scipy.integrade.ode – equivalent to Matlab's ode45

```python
def func_chem(t,X,k1=0.04,k2=10.0,k3=1.5e3):
    '''
    Function (RHS) of the system in problem 2:
    chemical reactions.
    '''
    F = np.zeros(X.size)
    F[0] = -k1*X[0] + k2*X[1]*X[2]
    F[1] = k1*X[0] - k2*X[1]*X[2] - 2.*k3*X[1]*X[1]
    F[2] = 2.*k3*X[1]*X[1]
    
    return F
```

```python
dt = .004
times = np.arange(0.,2500,dt)
ode45 = sp.integrate.ode(func_chem).set_integrator('dopri5')
ode45.set_initial_value(C0, 0.)
C = np.zeros((3,times.size))
C[:,0] = C0
ti = time.time()
for i in range(1,times.size):
    ode45.integrate(ode45.t+dt)
    C[:,i] = ode45.y
etime = time.time()-ti
print "ode45 integration executed in "  + str(etime) + " s"
```

```python
def my_odes(F,J,X0,t0,tf,dt,scheme='rk3'):
    
    '''
    Integrates dX/dt = F(X,t), where X and F 
    are N-dimensional arrays using RK3. 
    J is the jacobian of F used to estimate
    the time step.
    - t0/tf : initial/final time
    - X0 : initial value
    - dt : time step
    
    returns solution (X) and time array (t)
                                        '''
    try:
        N=X0.size
    except AttributeError:
        N=1
    
    X,Xold = X0,X0
    t,time = t0,t0
    i = 0
    
    while time <= tf:   
        Jt = J(Xold)
        lam = np.linalg.eigvals(Jt)
        lam_abs = np.abs(lam)
        dt = 0.75*(2.51/lam_abs.max())
        Xold = stepforward(Xold,F,time,dt,scheme)
        X = np.concatenate((X,Xold),axis=0)
        
        time = time + dt
        t = np.append(t,time)

    # reshape to  C0.size by t.size    
    X = X.reshape(X.size/C0.size,C0.size).T
        
    return X,t
```
